```python
import pandas as pd
import sqlite3
```

```python
#reading in the csv file and putting it into a dataframe
file_path = 'Resources/diabetes_prediction_dataset.csv'

# Use pandas to read the CSV file into a DataFrame
diabetes_df = pd.read_csv(file_path)
diabetes_df.head()
```

```python
#adding new column with binned age
bins = [0,12,18,35,65,80]
labels= ['children','teen','young adult','adult','senior']
diabetes_df['binned_age'] = pd.cut(diabetes_df['age'],bins=bins,labels=labels)
diabetes_df
```

```python
#dropping null values
refined_diabetes_df = diabetes_df[diabetes_df != 'No Info'].dropna()
```

```python
#cleaning up typos - changing ever to never
refined_diabetes_df['smoking_history'] = refined_diabetes_df['smoking_history'].replace('ever', 'never')
```

```python
#verifying value counts
refined_diabetes_df['smoking_history'].value_counts()
```

```python
# Creating an SQLite database file
database_name = 'diabetes_info.db'
conn = sqlite3.connect(database_name)
```

```python
#Exporting DataFrame to the SQLite database
refined_diabetes_df.to_sql('diabedes_prediction', conn, index=False, if_exists='replace')
```

```python
conn.close()
```
